```python
import os
import shutil
import xml.etree.ElementTree as ET
from tqdm import tqdm

# Function to convert bounding box coordinates to YOLO format
def convert_to_yolo(size, box):
    dw = 1.0 / size[0]
    dh = 1.0 / size[1]
    x = (box[0] + box[1]) / 2.0
    y = (box[2] + box[3]) / 2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x * dw
    w = w * dw
    y = y * dh
    h = h * dh
    return (x, y, w, h)

dataset_path = "/diagram/yolov8_dl/train"
kaggle_dataset_path = "/diagram/yolov8_dl//working/dataset"
os.makedirs(kaggle_dataset_path, exist_ok=True)

# Create train and test folders inside the dataset folder
train_path = os.path.join(kaggle_dataset_path, "train")
os.makedirs(train_path, exist_ok=True)
train_images_path = os.path.join(train_path, "images")
os.makedirs(train_images_path, exist_ok=True)
train_labels_path = os.path.join(train_path, "labels")
os.makedirs(train_labels_path, exist_ok=True)

image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.gif')  # Thêm bất kỳ định dạng nào bạn muốn hỗ trợ
image_files = [f for f in os.listdir(dataset_path) if f.lower().endswith(image_extensions)]

# Copy images and convert XML to YOLO labels
for image_file in tqdm(image_files):
    image_path = os.path.join(dataset_path, image_file)
    label_file = os.path.splitext(image_file)[0] + ".xml"
    label_path = os.path.join(dataset_path, label_file)
    
    if os.path.exists(image_path) and os.path.exists(label_path):
        # Copy the image
        shutil.copy(image_path, train_images_path)
        
        # Parse the XML file
        tree = ET.parse(label_path)
        root = tree.getroot()
        size = root.find("size")
        width = int(size.find("width").text)
        height = int(size.find("height").text)
        
        # Create a new YOLO format label file
        yolo_label_file = os.path.join(train_labels_path, os.path.splitext(image_file)[0] + ".txt")
        with open(yolo_label_file, "w") as f:
            for obj in root.iter("object"):
                class_name = obj.find("name").text
                bbox = obj.find("bndbox")
                xmin = float(bbox.find("xmin").text)
                xmax = float(bbox.find("xmax").text)
                ymin = float(bbox.find("ymin").text)
                ymax = float(bbox.find("ymax").text)
                x, y, w, h = convert_to_yolo((width, height), (xmin, xmax, ymin, ymax))
                f.write(f"{0} {float(x)} {float(y)} {float(w)} {float(h)}\n")
```

```python
import os
import cv2
from tqdm import tqdm
import matplotlib.pyplot as plt
dataset_path = "/diagram/yolov8_dl/working"
# Path to the folder containing the images with bounding box labels
images_path = "/diagram/yolov8_dl//working/dataset/train/images"
labels_path = "/diagram/yolov8_dl//working/dataset/train/labels"

image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.gif')  # Thêm bất kỳ định dạng nào bạn muốn hỗ trợ
image_files = [f for f in os.listdir(dataset_path) if f.lower().endswith(image_extensions)]

# Function to draw bounding boxes on images
def draw_boxes(image, label_file):
    with open(label_file, "r") as f:
        for line in f:
            class_name, x, y, w, h = line.strip().split()
            x, y, w, h = float(x), float(y), float(w), float(h)
            x1 = int((x - w / 2) * image.shape[1])
            y1 = int((y - h / 2) * image.shape[0])
            x2 = int((x + w / 2) * image.shape[1])
            y2 = int((y + h / 2) * image.shape[0])
            cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)
            cv2.putText(image, class_name, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
i=0
# Display images with bounding boxes
for image_file in tqdm(image_files):
    image_path = os.path.join(images_path, image_file)
    label_file = os.path.join(labels_path, os.path.splitext(image_file)[0] + ".txt")
    
    # Read the image
    image = cv2.imread(image_path)
    if image is None:
        continue
    
    # Draw bounding boxes
    draw_boxes(image, label_file)
    
    # Display the image using plt
    plt.figure(figsize=(10, 10))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
    if i==3:
        break
    i=i+1
```

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Tải mô hình đã huấn luyện
model = YOLO(r'C:\diagram\yolov8_dl\working\dataset\runs\detect\train\weights\best.pt')

# Đường dẫn đến tệp hình ảnh
image_path = r'C:\diagram\yolov8_dl\1.jpg'

# Dự đoán trên một hình ảnh
results = model(image_path)

# Đọc hình ảnh
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Vẽ bounding box lên hình ảnh
for result in results:
    boxes = result.boxes
    for box in boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # Tọa độ bounding box
        confidence = box.conf[0].item()  # Độ tin cậy
        class_id = box.cls[0].item()  # ID lớp
        
        # Vẽ bounding box
        cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)
        # Vẽ nhãn
        label = f'Class: {int(class_id)}, Conf: {confidence:.2f}'
        cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

# Hiển thị hình ảnh với bounding box
plt.imshow(img)
plt.axis('off')  # Tắt trục
plt.show()
```
